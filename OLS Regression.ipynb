# preprocessing

```python
#handling missing values
```

```python
df.isnull().sum()
```

```python
sns.boxplot(y=df['Avg. Area Income'])
```

```python
df['Avg. Area Income']=df['Avg. Area Income'].fillna(df['Avg. Area Income'].median())
```

```python
sns.boxplot(y=df['Avg. Area Number of Rooms'])
```

```python
df['Avg. Area Number of Rooms']=df['Avg. Area Number of Rooms'].fillna(df['Avg. Area Number of Rooms'].median())
```

```python
sns.boxplot(y=df['Avg. Area Number of Bedrooms'])
```

```python
df['Avg. Area Number of Bedrooms']=df['Avg. Area Number of Bedrooms'].fillna(df['Avg. Area Number of Bedrooms'].mean())
```

```python
df.isnull().sum()
#all the missing values have been handled
```

```python
#dropping the variable address since it is not of significance
df_new=df.drop(['Address'],axis=1)
df_new.head()
```

```python
#splitting into independent and dependent variables
x=df_new.iloc[:,0:5]
x.head()
```

```python
y=df['Price']
y.head()
```

# Splitting into training and testing data

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.75,random_state=101)
```

```python
import statsmodels.regression.linear_model as sm
from statsmodels.regression.linear_model import OLS
```

```python
model=sm.OLS(endog=ytrain,exog=xtrain).fit()
model
```
